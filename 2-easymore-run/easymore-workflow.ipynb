# Run `easymore`

`easymore` is a collection of functions that allows extraction of the data from a NetCDF file for a given shapefile such as a basin, catchment, points or lines. It can map gridded data or model output to any given shapefile and provide area average for a target variable.

Given the outputs of `datatool`, we can calculate the average of intersecting grid cells of the `RDRSv2.1` dataset with each computatioal element of the `Bow River at Calgary` watershed.

```python
import os

from easymore import Easymore
```

```python
# define the home directory
home=os.getenv('HOME')

# define settings for `easymore'
settings = {
    "case_name": "averaged",
    "temp_dir": f"{home}/scratch/cache-easymore/",
    "target_shp": f"{home}/scratch/alliancecan-training-data/modelling/0-geofabric/bow-at-calgary-geofabric/bcalgary_subbasins.shp",
    "target_shp_ID": "COMID",
    "source_nc": f"{home}/scratch/alliancecan-training-data/modelling/1-outputs/datatool-outputs/**/*.nc",
    "var_lon": "lon",
    "var_lat": "lat",
    "var_names": [
    "RDRS_v2.1_P_P0_SFC",
    "RDRS_v2.1_P_HU_09944",
    "RDRS_v2.1_P_TT_09944",
    "RDRS_v2.1_P_UVC_09944",
    "RDRS_v2.1_A_PR0_SFC",
    "RDRS_v2.1_P_FB_SFC",
    "RDRS_v2.1_P_FI_SFC"
    ],  
    "remapped_var_id": "COMID",
    "remapped_dim_id": "COMID",
    "output_dir": f"{home}/scratch/alliancecan-training-data/modelling/1-outputs/easymore-outputs/",
    "parallel": False,
}
bow_river_experiment = Easymore(**settings)

# running the experiment
bow_river_experiment.nc_remapper()
```
